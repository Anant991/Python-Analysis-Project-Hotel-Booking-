```python
# In Resort Hotel how much percentage of reservations are cancelled and how much not cancelled

resort_hotel=csv_1[csv_1["hotel"]=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
# In City Hotel how much percentage of reservations are cancelled and how much not cancelled

city_hotel=csv_1[csv_1["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
# Check for effect of price in both the hotels

resort_hotel=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[["adr"]].mean()

# Beacuse ek hi din main bahut data/records hain so us din ke average daily rate ko hum uska mean value lenge
```

```python
plt.figure(figsize=(20,8))
plt.title("Average Daily rates in City and Resort Hotel",fontsize=30)
plt.plot(resort_hotel.index,resort_hotel["adr"],label="Resort Hotel")
plt.plot(city_hotel.index,city_hotel["adr"],label="City Hotel")

plt.legend(fontsize=20)

plt.show()
```

```python
# So City Hotel prices as compared to the Resort Hotel is less
```

```python
# Check for months in which the cancellation is high

csv_1["month"]=csv_1["reservation_status_date"].dt.month
# Because hamre pass month column nhi h,so hum "reservation_status_date" ko hmne "datetime" main convert kiya tha, se month extract kr lenge

plt.figure(figsize=(16,8))
ax1=sns.countplot(x="month",hue="is_canceled",data=csv_1,palette="bright")

plt.title("Reservation Status Per Month",size=21)
plt.xlabel("Month")
plt.ylabel("Number of Reservations")

plt.legend(["not canceled","canceled"])

plt.show()
```

```python
# So result showing is highest cancellation is in January month and lowest in August month
# Number of reservations are highest in August and lowest in January
```

```python
# Extra

plt.figure(figsize=(16,8))
ax1=sns.countplot(x="arrival_date_month",hue="is_canceled",data=csv_1,palette="bright")

plt.title("Reservation Status Per Month",size=21)
plt.xlabel("Month")
plt.ylabel("Number of Reservations")

plt.legend(["not canceled","canceled"])

plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR per month",fontsize=21)
sns.barplot("month","adr",data=csv_1[csv_1["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index())
plt.legend(fontsize=20)
plt.show()
```

```python
# So The barplot show the "adr" is lowest in August month and Also we saw in the upper countplot the cancellations are minimum in August month
# And "adr" is highest in January and cancellations are also high in Jnuary month
```

```python
# Check for cancellations based on Country wise (only Top 10 Countries)

cancelled_data=csv_1[csv_1["is_canceled"]==1]
top_10_country=cancelled_data["country"].value_counts()[:10]

plt.figure(figsize=(8,8))
plt.title("Top 10 Contries with Reservation Cancelled")
plt.pie(top_10_country,autopct="%.2f",labels=top_10_country.index)
plt.show()
```

```python
# So The highest cancellations in Portugal Country around 70 %
```

```python
# So Check for the price of rooms which are cancelled or not cancelled

cancelled_csv1_adr=cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
cancelled_csv1_adr.reset_index(inplace=True)
cancelled_csv1_adr.sort_values("reservation_status_date",inplace=True)

not_cancelled_data=csv_1[csv_1["is_canceled"]==0]
not_cancelled_csv1_adr=not_cancelled_data.groupby("reservation_status_date")[["adr"]].mean()
not_cancelled_csv1_adr.reset_index(inplace=True)
not_cancelled_csv1_adr.sort_values("reservation_status_date",inplace=True)

plt.figure(figsize=(20,6))
plt.title("Average Daily Rate_1",fontsize=30)
plt.plot(cancelled_csv1_adr["reservation_status_date"],cancelled_csv1_adr["adr"],label="cancelled")
plt.plot(not_cancelled_csv1_adr["reservation_status_date"],not_cancelled_csv1_adr["adr"],label="not cancelled")
plt.legend(fontsize=20)
plt.show()
```

```python
# We have very hasy data, we saw in plot we have regular data between 2016 to 2017, So we plot that data

cancelled_csv1_adr=cancelled_csv1_adr[(cancelled_csv1_adr["reservation_status_date"]>"2016-01") & (cancelled_csv1_adr["reservation_status_date"]<"2017-09")]
not_cancelled_csv1_adr=not_cancelled_csv1_adr[(not_cancelled_csv1_adr["reservation_status_date"]>"2016") & (not_cancelled_csv1_adr["reservation_status_date"]<"2017-09")]

plt.figure(figsize=(20,6))
plt.title("Average Daily Rate_2",fontsize=30)
plt.plot(cancelled_csv1_adr["reservation_status_date"],cancelled_csv1_adr["adr"],label="cancelled")
plt.plot(not_cancelled_csv1_adr["reservation_status_date"],not_cancelled_csv1_adr["adr"],label="not cancelled")
plt.legend(fontsize=15)
plt.show()
```

```python
# In the last plot the cancellation line (Which is blue in color) is higher than non_cancellation line(Which is orange in color)
# So That proves the Price of Room is also affecting the cancellation
```
